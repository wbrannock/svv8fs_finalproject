- Description of TFIDIF formula ($\LaTeX$ OK):

I'm using the tf_method='double_norm', t_norm_k=0.5, and idf_method='standard' so we get the formula below after putting everything together. 

  $$
  \text{tfidf}_{t,d}
  =
  \left(
  0.5 + \frac{n_{t,d} + 0.5}{\max_{t' \in d}(n_{t',d}) + 0.5}
  \right)
  \log_2\left(\frac{N}{df_t}\right)
  $$

  where $n_{t,d}$ is the count of term $t$ in document $d$, $N$ is the total number of
  documents, and $df_t$ is the number of documents containing term $t$.
